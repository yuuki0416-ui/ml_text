#**fanction 復習**

Newton method を題材にしたコード

```python
import numpy as np
import matplotlib.pyplot as plt

# 1. 関数 f(x) を定義
def f(x):
    """
    f(x) = x^3 - 2x^2 + x + 3 を計算する関数
    """
    return x**3 - 2*x**2 + x + 3

# 2. xの値の範囲を設定
# -1.0 から 2.5 の範囲で、等間隔に100点のx座標を生成します
x = np.linspace(-1.0, 2.5, 100)

# 3. 対応するyの値を計算
y = f(x)

# 4. グラフの描画設定
plt.figure(figsize=(8, 6)) # グラフのサイズを設定
plt.plot(x, y, label=r'$f(x) = x^3 - 2x^2 + x + 3$', color='blue') # グラフをプロット

# 5. グラフの装飾
plt.title('Graph of $f(x) = x^3 - 2x^2 + x + 3$') # タイトル
plt.xlabel('$x$') # x軸ラベル
plt.ylabel('$f(x)$') # y軸ラベル
plt.grid(True, linestyle='--', alpha=0.6) # グリッド線を表示
plt.axhline(0, color='black', linewidth=0.5) # x軸 (y=0) を描画
plt.axvline(0, color='black', linewidth=0.5) # y軸 (x=0) を描画
plt.legend() # 凡例を表示

# 6. グラフの表示
plt.show()

# グラフのイメージ（実行結果）
#
```
