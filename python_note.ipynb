#**mapの使い方**
   
   
    list(map(関数,  イテラブル))


   **iterable**とは

   要素を1つずつ取り出して反復処理できるオブジェクトのこと

   list

   tuple
   
   str
   
   dict
  
  など

```python
# --- 入力部分（変更なし）---
N = int(input())
X = int(input())
A = list(map(int, input().split()))

# --- 処理と出力部分（追加）---
Answer = False

# リストAを全探索する
for i in range(N):
    if A[i] == X:
        Answer = True
        # 一度見つかれば、もう探す必要はないのでループを抜ける（効率化）
        break

# 結果の出力
# AnswerがTrueの場合に実行される
if Answer:
    print("Yes")
# AnswerがFalseの場合に実行される
else:
    print("No")


```
